## Read Data
Read splited raw datasets, clean header, rplace nan, and concat through time.

```python
# concat FINFSTA

dir_path = "./../data/raw/FQ/FINFSTA_quarterly_consolidated"

# get all paths in selected directody
file_paths = os.listdir(dir_path)

sheet_name = '決算期ベース.有価証券報告書優先.連結優先(IFRS優先)'

df = pd.DataFrame()
for path in file_paths:
    # Define path
    data_file_path = dir_path + "/" + path
    
    # read temporary data file
    temp = Dataset("xlsx", sheet_name, data_file_path)
    temp.clean_header()
    
    # concat
    df = pd.concat([df, temp.data], axis=0)
    print(path + " concated")
```

```python
# save this temporary df as csv
df.reset_index(drop=True, inplace=True)
df.to_csv("./../data/processed/tse1_concated.csv")
```

## Multiindexing Panel data with firm
https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html <br>
https://python.quantecon.org/pandas_panel.html <br>
https://deepage.net/features/pandas-multiindex.html

```python
df.columns
```

```python
# drop unused column
df.drop(['Unnamed: 1', 'Unnamed: 2'], axis=1, inplace=True)
```

```python
# rename firm name and fiscal period columns
df.rename(columns={'Unnamed: 0.1': '企業名'}, inplace=True)
df.rename(columns={'Unnamed: 3': '決算期'}, inplace=True)
```

```python
# change date columns from str to datetime
# 決算期は月データ。
# https://teratail.com/questions/280394
# https://stackoverflow.com/questions/14425133/date-object-with-year-and-month-only
for col in ['決算発表日', '事業年度開始年月日［３ヵ月］', '事業年度終了年月日']:
    df[col] = pd.to_datetime(df[col], yearfirst=True, format="%Y/%m/%d")

# df["決算期"] = df["決算期"].apply(lambda x: (int(x[:4]), int(x[4+1:])))
df
```

```python
# generate year and month column for "決算期"
df["決算期（年）"] = df["決算期"].apply(lambda x: int(x[:4]))
df["決算期（月）"] = df["決算期"].apply(lambda x: int(x[4+1:]))
df
```

```python
# set multiindex ["企業名", "決算期"]
df = df.set_index(["企業名", "決算期（年）", "決算期（月）"])

# sorting firm name for multiindex slicing
df.sort_index(level=0, inplace=True)

df
```
