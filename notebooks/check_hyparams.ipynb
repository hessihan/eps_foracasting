```python
ai = pd.read_csv("../assets/y_hats/accuracy_table_i.csv", index_col=[0, 1])
y_hat = pd.read_csv("../assets/y_hats/multivariate/y_hat_mraf_i_tuned_simple.csv", index_col=[0, 1, 2])
```

```python
good_firm = ai["y_hat_mraf_i_tuned_simple"].loc[pd.IndexSlice[:, "MAPE-UB"]].sort_values()[:100].index
plt.figure(figsize=(16, 9))
plt.hist(y_hat.loc[pd.IndexSlice[good_firm, 2018, "Q1"]]["best_hyparam"])
plt.title("best hyparam of " + y_hat.columns[0])
plt.xticks(rotation=90)
plt.show()
```

```python
bad_firm = ai["y_hat_mraf_i_tuned_simple"].loc[pd.IndexSlice[:, "MAPE-UB"]].sort_values()[-100:].index
plt.figure(figsize=(16, 9))
plt.hist(y_hat.loc[pd.IndexSlice[bad_firm, 2018, "Q1"]]["best_hyparam"])
plt.title("best hyparam of " + y_hat.columns[0])
plt.xticks(rotation=90)
plt.show()
```
