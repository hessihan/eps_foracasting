```python
import pandas as pd
import sys
sys.path.insert(1, '../src')
from utils.data_editor import lag
```

```python
col = ['EPS', 'INV', 'AR', 'CAPX','GM', 'SA', 'ETR', 'LF']
df = pd.read_csv("../data/processed/tidy_df.csv", index_col=[0, 1, 2])
df = df[col]
df
```

```python
def lagged_data(n):
    lag_n = df.groupby(level=0).shift(n)
    lag_n.columns = [i + "_-" + str(n) for i in col]
    return lag_n
```

```python
df_lagged = pd.concat([
    df,
    lagged_data(1),
    lagged_data(2),
    lagged_data(3),
    lagged_data(4),
    lagged_data(5),
    lagged_data(6),
    lagged_data(7),
    lagged_data(8),
    lagged_data(9),
], axis=1)
```

```python
df_lagged.to_csv("../data/processed/tidy_df_lag.csv")
```
